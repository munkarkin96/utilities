## Pool

```python
a = list(range(0, 100000))
```

```python
def map_fn(data):
    for i in range(10000):
        data += 1 
    return data 
```

```python
p = mp.Pool(mp.cpu_count())

start_time = time.time()
ret = p.map(map_fn, a)
end_time = time.time()

print("Took: {} s".format(end_time - start_time))
```

## Apply

```python
A = list(range(0, 100000))

import multiprocessing as mp
import numpy as np 
import time 

def map_fn(data):
    for i in range(10000):
        data += 1 
    return data 

p = mp.Pool(mp.cpu_count())

start_time = time.time()
ret = [p.apply(map_fn, args=(a, )) for a in A] #_async
end_time = time.time()

print("Took: {} s".format(end_time - start_time))
```

## Pool.apply_sync

The Pool.map and Pool.apply will lock the main program until all processes are finished, which is quite useful if we want to obtain results in a particular order for certain applications. 
In contrast, the async variants will submit all processes at once and retrieve the results as soon as they are finished. One more difference is that we need to use the get method after the apply_async() call in order to obtain the return values of the finished processes.

```python
A = list(range(0, 100000))

import multiprocessing as mp
import numpy as np 
import time 

def map_fn(data):
    for i in range(10000):
        data += 1 
    return data 

p = mp.Pool(mp.cpu_count())

start_time = time.time()
ret = [p.apply_async(map_fn, args=(a, )) for a in A] #_async
end_time = time.time()

print("Took: {} s".format(end_time - start_time))
```

```python
output = [p.get() for p in ret]

output 
```

## Map Reduce in Batches

```python
import multiprocessing as mp 
import time 
A = list(range(0, 100000))

def map_fn(batch, common_list):
    
    for data in batch:

        for i in range(10000):
            data += 1 
        
        common_list.append(data)

start_time = time.time()


split_A = np.array_split(np.array(A), mp.cpu_count())



process = []
m = mp.Manager() 
common_list = m.list() 

for batch in split_A:
    process.append(mp.Process(target = map_fn, args = (batch, common_list)))
    
for p in process:
    p.start()
    
for p in process:
    p.join() 
    
end_time = time.time()

print("Took: {} s".format(end_time - start_time))
```

## MultiThreading

```python
A = list(range(0, 100000))
```

```python
import time
import threading

A = list(range(0, 100000))

ret = [] 
def map_fn(data):
    for i in range(10000):
        data += 1 
    # return data 
    ret.append(data)

start_time = time.time() 
num_threads = 5 

for i, data in enumerate(A): 
    threads =[] 
    
    try:
        feed = A[i:i+num_threads]
    except:
        continue 

    for i in range(num_threads):
        threads.append(threading.Thread(target=map_fn, args = (feed[i], ) ))

    for t in threads:
        t.start()

    for t in threads:
        t.join()
    
end_time = time.time() 

print("Took: {} s".format(end_time - start_time))
```
